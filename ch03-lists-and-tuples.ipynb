## 10. 參數加總

完成以下 `my_sum()` 函式，其輸入為不定個數且相同型態的參數 (如一堆數字或一堆字串)，輸出為所有參數的加總 (使用 + 運算子)；若沒有傳入參數則回傳 None

````python
def mysum(*elements):
    output = None
    if not elements:
        return output

    output = elements[0]
    for e in elements[1:]:
        output += e
    return output

print(mysum(1, 2, 3))
print(mysum("Hello", " ", "World", "!"))
print(mysum([1], [2, 3], [4, 5, 6]))
print(mysum((7, 6), (5, 4, 3), (2, 1)))
print(mysum())
````

## 11. 多重排序 (一)

完成以下 `sort_by_age()` 函式，其輸入為一個 list of dicts, 每個 dict 包含了姓名 (name) 與年齡 (age), 請將輸入依照年齡由小到大排序後回傳，若年齡相同，則再依照姓名的字典順序排序

````python
def sort_by_age(people):
    return sorted(people, key=lambda x: (x["age"], x["name"]))

people = [
    {"name": "Doe", "age": 30},
    {"name": "Doe", "age": 25},
    {"name": "Cooper", "age": 30},
    {"name": "Baker", "age": 22}
]

for person in sort_by_age(people):
    print(person)
````

排序是實務上常見的操作，關於 `sorted()` 函式的一些重點如下：

1. `sorted()` 會回傳排序好的序列 (預設順序為由小到大)，但並不會改變序列本身

```
my_list = [9, 3, 1, 7, 6]
new_list = sorted(my_list)
print(my_list)  # [9, 3, 1, 7, 6], my_list 不會改變
print(new_list)  # [1, 3, 6, 7, 9]
```

2. `sorted()` 用來排序字串 list 時，預設順序為字典順序，但是可以使用 `key` 參數指定排序方式

```
words = "beautiful is better than ugly".split()
sorted(words)  # ["beautiful", "better", "is", "than", "ugly"]
# 使用 len() 決定字串順序
sorted(words, key=len)  # ["is", "than", "ugly", "better", "beautiful"]
```

3. 常見的自訂 sort key 寫法是使用匿名函式 (lambda), lambda 函式是長度較短，你不想為它特別取名的函式。如上述解答中，`sorted(people, key=lambda x: (x["age"], x["name"])`, 變數 `x` 指的是序列裡面的元素 (`people` 裡面的每一個 dict), 我們指定使用 `x["age"]` (age 這個 key 的值) 為第一個排序標準，`x["name"]` 為次要排序標準


## 12. 出現最多次的城市

完成以下 `most_common_cities()` 函式，其輸入為城市名稱列表 (包含重複值), 輸出為列表中出現最多次的城市名


````python
from collections import Counter

def most_common_cities(cities):
    return Counter(cities).most_common(1)[0][0]

cities = ["高雄市", "台北市", "新竹市", "新竹市", "彰化縣", "高雄市", "台北市", "彰化縣", "台北市", "新竹市", "台北市", "台北市", "高雄市"]
print(most_common_cities(cities))
````

`collections.Counter` 是計算一個序列中元素出現頻率的常用方法。其輸入為一個序列 (list, tuple, strings), 其輸出可以想成是 dict, 內容包含了各個元素及其出現次數。`.most_common()` 方法會將元素依照出現頻率排序 (如果傳入參數 n, 代表只看前 n 名)，其內容為一個 list of tuples. 如以下範例

````python
from collections import Counter

cities = ["高雄市", "台北市", "新竹市", "新竹市", "彰化縣", "高雄市", "台北市", "彰化縣", "台北市", "新竹市", "台北市", "台北市", "高雄市"]
freqency = Counter(cities)

print(freqency)  # Counter({'台北市': 5, '高雄市': 3, '新竹市': 3, '彰化縣': 2})
print(freqency.most_common())  # [('台北市', 5), ('高雄市', 3), ('新竹市', 3), ('彰化縣', 2)]

# 可以把 frequency 當成 dict 操作
for k, v in freqency.items():
    print(k, v)
````

## 13. 序列合併

完成以下 `combine_lists()` 函式，其輸入為兩個 lists, 列表中的元素是 (key, value) 的 tuple, 輸出這兩個 lists 合併後的結果

````python
def combine_lists(list_a, list_b):
    combined_dict = {}

    for key, value in list_a:
        combined_dict[key] = combined_dict.get(key, 0) + value

    for key, value in list_b:
        combined_dict[key] = combined_dict.get(key, 0) + value

    return [(key, value) for key, value in combined_dict.items()]

list_a = [('A', 1), ('B', 2), ('C', 3)]
list_b = [('B', 3), ('C', 4), ('D', 5)]

print(combine_lists(list_a, list_b))  # [('A', 1), ('B', 5), ('C', 7), ('D', 5)]
````

此處我們用一個 dict 紀錄兩個 list 中出現過的 key 與 value. `.get(KEY, DEFAULT_VALUE)` 方法可以取得一個 dict 中 KEY 的值，若 KEY 不存在則回傳 DEFAULT_VALUE
